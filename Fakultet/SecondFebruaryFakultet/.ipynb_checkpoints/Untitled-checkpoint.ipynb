```python
def calc(x, a, b):
    s = ""
    x = int(str(x), a)
    while x > 0:
        s += str(x % b)
        x //= b
    return s[::-1]


for i in range(100):
    if (len(calc(i, 10, 6)) == 2) and (len(calc(i, 10, 5)) == 3) and (calc(i, 10, 11).endswith("1")):
        print(i)
```

```python
arr = [int(x) for x in open("17.txt").readlines()]
odds = list(filter(lambda x: x % 2 != 0, arr))
mid = sum(odds) / len(odds)
maxx = -10000000000
count = 0
for i in range(len(arr) - 1):
    a = arr[i]
    b = arr[i+1]
    if ((a % 5 == 0) or (b % 5 == 0)) and ((a < mid) or (b < mid)):
        count += 1
        if a + b > maxx:
            maxx = a + b

count, maxx
```

```python
from functools import lru_cache
from math import ceil

def moves(h):
    a, b = h
    return (a - 1, b), (a, b - 1), (ceil(a / 2), b), (a, ceil(b / 2))


@lru_cache(None)
def game(h):
    if h[0] <= 1 or h[1] <= 1:
        return
    if sum(h) <= 40:
        return "W"
    if any(game(m) == "W" for m in moves(h)):
        return "P1"
    if all(game(m) == "P1" for m in moves(h)):
        return "B1"
    if any(game(m) == "B1" for m in moves(h)):
        return "P2"
    if all(game(m) == "P1" or game(m) == "P2" for m in moves(h)):
        return "B1/2"

for s in range(21, 100):
    if game((20, s)) is not None:
        print(s, game((20, s)))
```

```python
def tr(start, finish):
    if start > finish:
        return 0
    if start == finish:
        return 1
    if start == 15:
        return 0
    return tr(start + 1, finish) + tr(start + 2, finish)

tr(3, 9) * tr(9, 20)
```

```python
def p(x):
    return x > 1 and all(x % i != 0 for i in range(2, int(x**0.5) + 1))


def div(x):
    d = set()
    for i in range(2, int(x**0.5)+1):
        if x % i == 0:
            d |= {i, x // i}
    return sorted(d)


for i in range(452_022, 1_000_000):
    if p(i):
        continue
    arr = div(i)
    M = arr[0] + arr[-1]
    if M % 7 == 3:
        print(i, M)
```

```python
f = open("27-A.txt")
n = int(f.readline())
print(n)
arr = [int(x) for x in f.readlines()]
minn = 10000000000
for i in range(n):
    a = arr[i]
    for j in range(i+1, n):
        b = arr[j]
        for k in range(j+1, n):
            c = arr[k]
            if (a+b+c) % 3 == 0:
                if (a+b+c) < minn:
                    minn = (a+b+c)
minn
```

```python
from itertools import combinations

f = open("27-B.txt")
n = int(f.readline())
a = [int(x) for x in f.readlines()]
a.sort()
a = a[:100]

m = float("inf")
for x1, x2, x3 in combinations(a, 3):
    if (x1+x2+x3)%3==0:
        m = min(x1+x2+x3, m)
print(m)
```
